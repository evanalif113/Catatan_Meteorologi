```python

from datetime import datetime
import meteostat as ms
import pandas as pd

# ==========================================
# 1. PARAMETER UTAMA
# ==========================================
# Gunakan WMO ID dalam bentuk String
wmo_id = '96805'  

# Tentukan rentang waktu (Contoh: 1 Januari 2024 - 31 Januari 2024)
year = 2025
start = datetime(year, 1, 1)
end = datetime(year, 12, 31, 23, 59)

print(f"Mengakses server untuk WMO ID: {wmo_id}...")

# ==========================================
# 2. EKSEKUSI PENARIKAN DATA
# ==========================================
# Kita langsung memasukkan wmo_id ke dalam fungsi Hourly
data = ms.hourly(ms.Station(wmo_id), start, end)  # Alternatif jika cara pertama gagal
df = data.fetch()
```

```python
# ==========================================
# 3. VALIDASI & SIMPAN KE CSV
# ==========================================
if df.empty:
    print("GAGAL: Data kosong atau server sedang gangguan untuk tanggal tersebut.")
else:
    # Mengisi data yang bolong (NaN) dengan teknik interpolasi garis lurus
    # (Sangat penting karena sensor WMO kadang mati selama 1-2 jam)
    df = df.interpolate() 
    
    # Simpan ke CSV
    file_name = f"dataset_wmo_{wmo_id}_{year}.csv"
    df.to_csv(file_name, index=True)
    
    print(f"SUKSES! Data berhasil diunduh dan disimpan sebagai: {file_name}")
```
