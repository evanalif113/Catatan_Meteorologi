```python
import math

def hitung_curah_hujan():
    print("=== 🌧️ Kalkulator Curah Hujan Penakar Manual ===")
    
    try:
        # Meminta input dari pengguna (bisa desimal)
        diameter_cm = float(input("Masukkan diameter penampang (cm) : "))
        volume_ml = float(input("Masukkan volume air (ml)         : "))
        
        # Proses perhitungan
        jari_jari = diameter_cm / 2
        luas_penampang = math.pi * (jari_jari ** 2)  # Luas dalam cm kuadrat
        
        # Volume (ml) itu sama dengan cm kubik. Jadi Tinggi = Volume / Luas
        curah_hujan_cm = volume_ml / luas_penampang
        
        # Konversi dari cm ke milimeter (standar BMKG)
        curah_hujan_mm = curah_hujan_cm * 10
        
        # Menampilkan hasil dengan 2 angka di belakang koma
        print("-" * 45)
        print(f"Hasil Perhitungan Curah Hujan  : {curah_hujan_mm:.2f} mm")
        print("-" * 45)
        
    except ValueError:
        print("Error: Mohon pastikan yang Anda masukkan adalah angka (gunakan titik untuk desimal, contoh: 8.5)")

# Menjalankan program
hitung_curah_hujan()
```
